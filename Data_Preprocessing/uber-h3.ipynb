```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import pandas as pd
import requests
import json
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import h3
import h3.api.basic_int as h3_int
import h3.api.numpy_int as h3_np
import h3.api.memview_int as h3_mv
from shapely.geometry import Point, Polygon
from geopy.distance import distance
import geopandas as gpd
import pandas as pd
import h5py
```

```python
for i in tqdm(range(1, 21)):
  gps = pd.read_csv(f'/content/drive/MyDrive/GPS/tri_bike_parts{i}.csv')
  arr = np.array(gps)
  temp = []
  temp_x = []
  temp_y = []

  for j in range(len(arr)):
    grid_9 = h3.geo_to_h3(arr[j,2], arr[j,3], 9)
    temp.append(grid_9)
    x, y = h3.h3_to_geo(grid_9)
    temp_x.append(x)
    temp_y.append(y)
    
  gps['h3_9'] = temp
  gps['h3_9_lat'] = temp_x
  gps['h3_9_lng'] = temp_y

  gps.to_csv(f'/content/drive/MyDrive/data/H3-9/tri_bike_parts{i}.csv')
```
