```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString
import numpy as np
from tqdm import tqdm
```

```python
data_all = pd.read_csv('/content/drive/MyDrive/data/bike/bike_gps1.csv')
```

```python
data = data_all.drop(['Unnamed: 0'], axis=1).iloc[:10000]
data = data.drop(['str_100m'], axis=1)
```

```python
data = data.dropna()
```

```python
idx = data['trackingNo'].tolist()
lon = data['longitude'].tolist()
lat = data['latitude'].tolist()
user = data['useId'].tolist()

lon_temp = []
lat_temp = []

lon_mean = []
lat_mean = []

count = 0

for i in tqdm(range(len(data))):
  if i == len(data) - 1:
    count += 1
    lon_temp.append(lon[i])
    lat_temp.append(lat[i])

    lon_mean.extend([np.mean(lon_temp) for i in range(count)])
    lat_mean.extend([np.mean(lat_temp) for i in range(count)])

  elif user[i] == user[i+1]:
    count += 1
    lon_temp.append(lon[i])
    lat_temp.append(lat[i])

  else:
    count += 1
    lon_temp.append(lon[i])
    lat_temp.append(lat[i])

    lon_mean.extend([np.mean(lon_temp) for i in range(count)])
    lat_mean.extend([np.mean(lat_temp) for i in range(count)])

    count = 0
    lon_temp = []
    lat_temp = []
```

```python
data['lat_mean'] = lat_mean
data['lon_mean'] = lon_mean
```

```python
idx = data['trackingNo'].tolist()
lon = data['longitude'].tolist()
lat = data['latitude'].tolist()
lon_mean = data['lon_mean'].tolist()
lat_mean = data['lat_mean'].tolist()

lon_temp = []
lat_temp = []

errorsum = 0
rad = []
count = 0

for i in tqdm(range(len(data))):
  if i == len(data) - 1:
    count += 1
    lon_temp.append((lon[i] - lon_mean[i]) ** 2)
    lat_temp.append((lat[i] - lat_mean[i]) ** 2)

    errorsum = sum(lon_temp) + sum(lat_temp)
    rad.extend([errorsum/count  for i in range(count)])

  elif user[i] == user[i+1]:
    count += 1
    lon_temp.append((lon[i] - lon_mean[i]) ** 2)
    lat_temp.append((lat[i] - lat_mean[i]) ** 2)

  else:
    count += 1
    lon_temp.append((lon[i] - lon_mean[i]) ** 2)
    lat_temp.append((lat[i] - lat_mean[i]) ** 2)

    errorsum = sum(lon_temp) + sum(lat_temp)
    rad.extend([np.sqrt(errorsum/count)  for i in range(count)])

    count = 0
    lon_temp = []
    lat_temp = []
```
